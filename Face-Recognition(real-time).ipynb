```python
import face_recognition
import dlib
import cv2
from PIL import Image
import numpy as np
import os
import pandas as pd
```

```python
def face_detect(frame):
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.3,5)
    return faces,gray
```

```python

def take_pic():
    video_cap=cv2.VideoCapture(0)
    count=0
    r=False
    name=input("Give name of the person ")
    user=int(input("Enter user no"))
    print("Look into camera")
    while True:
        ref,frame=video_cap.read()
#         frame = cv2.flip(frame, -1)
        if(ref):
            faces,gray=face_detect(frame)
            img_name = "facedata/{}.{}.{}.jpg".format(user,name,count)
            if r==True:
                for (x,y,x1,y1) in faces:    

                    cv2.imwrite(img_name,gray[y:y1+y,x:x1+x])

                    print("{} read the image !".format(img_name))  
                    cv2.rectangle(frame,(x,y),(x1+x,y1+y),(154,224,255),2)
                    count+=1
        cv2.imshow("Video Feed",frame)
        ch=0xFF & cv2.waitKey(1)   
        
        if ch==ord("q"):
            break
        elif count==30:
            break
        elif ch==ord("r"):
            r=True
    video_cap.release()
    cv2.destroyAllWindows()
```

```python
def getIMGlabels(path):
    
    imgpath=[os.path.join(path,f) for f in os.listdir(path)]
    faceSam=[]
    ids=[]  
    names=set()
    for img in imgpath:
        fc=Image.open(img)
        ids.append(int((img.split(".")[0]).split("\\")[1]))
        names.add(img.split(".")[1])
        faceSam.append(np.asarray(fc))
    return faceSam,names,ids
       
```

```python

def detect():
    path="facedata"
    recognizer=cv2.face.LBPHFaceRecognizer_create()
    df=pd.DataFrame(columns=["names"])
    faces,names,ids=getIMGlabels(path) 
    recognizer.train(faces,np.array(ids))
    df["names"]=list(names)
    
    df.to_csv("facetrainer/names.csv",encoding="utf-8",header=True)
    recognizer.write('facetrainer/trainer.yml')
    print("\n [INFO] {} faces trained".format(len(names)))
```

```python

def pred():
    detect()
    recog=cv2.face.LBPHFaceRecognizer_create()
    recog.read('facetrainer/trainer.yml')
    df=pd.read_csv("facetrainer/names.csv")
    names=df["names"]
    video_cap=cv2.VideoCapture(0)
    font = cv2.FONT_HERSHEY_SIMPLEX 
    while True:
        ref,frame=video_cap.read()
#         frame = cv2.flip(frame, -1)
        if(ref):
            faces,gray=face_detect(frame)
            for (x,y,x1,y1) in faces:
                cv2.rectangle(frame,(x,y),(x1+x,y1+y),(154,224,255),2)
                
                id,confidence = recog.predict(gray[y:y1+y,x:x1+x])
                if(confidence < 100):
                    name=names[id]
                    confidence = "  {0}%".format(round(100 - confidence))
                else:
                    name = "unknown"
                    confidence = "  {0}%".format(round(100 - confidence))

                cv2.putText(frame,name,(x-5,y-5),font,1,(255,255,255),2)
                cv2.putText(frame,str(confidence),(x+5,y-5+y1),font,1,(255,255,255),2)
                         
        cv2.imshow("Video Feed",frame)
        ch=0xFF & cv2.waitKey(1)   
        
        if ch==ord("q"):
            break
    print("\n [INFO] Exiting Program and cleanup stuff ")
    video_cap.release()
    cv2.destroyAllWindows()
    
```

```python
c=True
while c:
    n=int(input("1 for new user,2 for detect ,3 for exit \n"))
    
    if n==3:
        c=False
        print("Thanks for using our app")
    elif n==1:     
        take_pic()
    elif n==2:
        print("Opening camera")
        pred()
```
